```python
from numpy import mean
mean(df[" Corrected Flux"])
np.percentile(df[" Corrected Flux"],q=5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

from scipy.misc import electrocardiogram
from scipy.signal import find_peaks
```

```python

x = df[" Corrected Flux"]
peaks, properties = find_peaks(-x,distance=160,width=0.2,rel_height=0.03,threshold=0.0001)
df["BJD - 2454833"].max()
print(np.std(x[peaks]))
plt.plot(x)
plt.plot(peaks, x[peaks], "x")
plt.show()
```

```python


# example data with peaks:
x = df["BJD - 2454833"]
data = df[" Corrected Flux"]




b = (np.diff(np.sign(np.diff(data))) > 0).nonzero()[0] + 1         # local min
b=list(b)
m=[]
for i in b:
    
    if data[i] <np.percentile(df[" Corrected Flux"],q=2):
        m.append(i)
    
# +1 due to the fact that diff reduces the original index number
print(m)
# plot
print(np.std(m))
plt.figure(figsize=(12, 5))
plt.plot(x, data, color='grey')


plt.plot(x[m], data[m], "o",  color='r')

plt.show()
```
